```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '../Manga109/kangaiset_annotation_long.csv'
df = pd.read_csv(file_path)
```

```python
genre_counts = df['emotion_str'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(genre_counts.index, genre_counts.values, color='blue', alpha=0.7, edgecolor='black')

plt.title('Bar Chart of Emotion Distribution')
plt.xlabel('Emotion')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Show the plot
plt.show()
```

```python
import seaborn as sns

# Create a pivot table to count the occurrences of emotions in each genre
pivot_table = df.pivot_table(index='book_genre', columns='emotion_str', aggfunc='size', fill_value=0)

# Create a heatmap to visualize the correlation
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='Blues', cbar=True)

# Customize the plot
plt.title('Heatmap of Emotions by Genre')
plt.xlabel('Emotion')
plt.ylabel('Genre')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Show the plot
plt.tight_layout()
plt.show()
```
