```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
df = pd.read_csv('manga_statistics.csv')

# Plotting the histogram of page count
plt.figure(figsize=(10, 6))
plt.hist(df['page_count'], bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Page Count')
plt.xlabel('Page Count')
plt.ylabel('Number of Manga Volumes')
plt.grid(axis='y', alpha=0.75)
plt.xticks(range(0, df['page_count'].max() + 1, 20))  # Adjust the x-ticks as needed
plt.show()
```

```python
# Count the occurrences of each genre
genre_counts = df['genre'].value_counts()

# Plotting the bar chart for genre distribution
plt.figure(figsize=(10, 6))
plt.bar(genre_counts.index, genre_counts.values, color='blue', alpha=0.7, edgecolor='black')

# Customize the plot
plt.title('Bar Chart of Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Number of Manga Volumes')
plt.xticks(rotation=45)  # Rotate x-ticks for better visibility
plt.grid(axis='y', alpha=0.75)

# Show the plot
plt.show()
```
