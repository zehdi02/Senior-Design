```python
from ultralytics import YOLO
import os

os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
model = YOLO("yolov8n_MangaVision.pt").to('cuda')
```

```python
result = model.train(
    data='MangaVision_dataset_sample/data.yaml',
    epochs=1,
    batch=-1,
    imgsz=640,
    save=True,
    device='cuda',
    verbose=True,
    amp=False,
    dropout=.05,
    val=True,
    plots=True
)
```

refer to:
https://docs.ultralytics.com/reference/utils/metrics/#ultralytics.utils.metrics.ConfusionMatrix.matrix

```python
print(dir(result))
```

```python
print(dir(result.box))
```

```python
# metrics to add on evaluation endpoint:

print(result.results_dict)
print()

for i in range(len(result.keys)):
    print(f'{result.keys[i]}: {result.maps[i]}')
print()

for i in range(len(result.box.ap_class_index)):
    print(f'class {i}:\n{result.box.all_ap[i]}')    # AP scores for all classes and all IoU thresholds
    print(f'ap50-95: {result.box.ap[i]}')
    print(f'ap50: {result.box.ap50[i]}\n')

print(f'mp: {result.box.mp}')
print(f'mr: {result.box.mr}\n')

print(result.speed)
print()

print(result.curves)
# print(result.curves_results)
print()

# confusion matrix metrics
print(result.confusion_matrix.matrix)   # confusion matrix
print(result.confusion_matrix.nc)   # The number of classes.
print(result.confusion_matrix.conf)   # confidence threshold for detections.
print(result.confusion_matrix.iou_thres)   # Intersection over Union threshold.
```
